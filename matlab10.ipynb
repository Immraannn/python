```python
import matplotlib.pyplot as plt
import numpy as np
t=np.arange(0,10,2)
plt.plot(t,t**2,'b+',label='^5')
plt.plot(t,t**2.2,'rs',label='^3')
plt.plot(t,t**2.5,'g^',label='^4')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=np.arange(-4,4,0.0001)
plt.plot(x,x**2,'ro',markersize=1)
plt.xlabel=['x']
plt.ylabel=['y']
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=np.arange(0,4,0.0001)
plt.plot(x,(x**2+4),'r--',markersize=1)
plt.xlabel=['x']
plt.ylabel=['y']
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,3,4,5,6]
y2=[0,9,8,7,6]
lines=plt.plot(x1,y1,x2,y2)#(line[0],line[1])
plt.setp(lines[0],color='r',markersize=1)
plt.setp(lines[1],color='g',markersize=1)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
subjects = ["Math", "Physics", "Chemistry", "CS", "English"]
marks = [85, 90, 78, 92, 88]

# Create bar graph
plt.bar(subjects, marks)

# Add labels and title
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Marks in Different Subjects")

# Show the graph
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
subjects = ["Math", "Physics", "Chemistry", "CS", "English"]
marks = [85, 90, 78, 92, 88]

# Create pie chart
plt.pie(marks, labels=subjects, autopct="%1.1f%%", startangle=90)

# Add title
plt.title("Marks Distribution")

# Show chart
plt.show()

```

```python
import matplotlib.pyplot as plt

# Example data (marks of students)
marks = [50, 55, 60, 65, 65, 70, 72, 75, 80, 82, 85, 85, 88, 90, 92, 95, 95, 98]

# Create histogram
plt.hist(marks, bins=8, color="skyblue", edgecolor="black")

# Labels and title
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Histogram of Marks Distribution")

# Show plot
plt.show()
```
