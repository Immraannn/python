```python
from sklearn.datasets import *      #Sklearn toy datasets 

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
```

```python
data = load_breast_cancer()
#data
#data.data
print(data.data.shape)      #x
print(data.target.shape)    #y ,data.target = output label (cancer or not)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data.data, data.target, train_size=0.9)

#tuned parameter for c(1/lambda) Hp
tp= [{'C': [10**-4,10**-2,10**0,10**2,10**4]}]

model= GridSearchCV(LogisticRegression(), tp, scoring='f1', cv=5) #GridSearch on top of Logistic Regressions
# GridSearchCV will:
#Try each value of C (or whatever you put in tp)
#Train Logistic Regression on 5 folds of data
#Compute F1-score for each parameter setting
#Select the best parameter that gives highest F1-score

model.fit(X_train, Y_train)    #training
#o this command:
#Runs Grid Search
#Trains Logistic Regression with each hyperparameter option
#Performs 5-fold cross-validation (cv=5)
#Selects the best parameter using F1 score
#Fits the final model
```

```python
#testing
print('model best parameter',model.best_estimator_)
print(model.score(X_test,Y_test))



```

```python
from sklearn.datasets import *      #Sklearn toy datasets 

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Predict on train & test
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Print accuracy
print("Train Accuracy:", accuracy_score(Y_train, y_pred_train))
print("Test Accuracy :", accuracy_score(Y_test, y_pred_test))




```

#### Expermenting Now with parameters
Playing with C to check sparcity

```python
import numpy as np
# Create Logistic Regression model with L1 regularization
# C = 0.1 → strong regularization (smaller C means stronger penalty)
# penalty='l1' → L1 regularization makes many weights zero (feature selection)
# solver='liblinear' → required solver for L1 penalty in Logistic Regression
clf= LogisticRegression(C=0.1, penalty='l1', solver='liblinear')  #solver is also imp param: Read Sklearn.LR docs

# Train the model on training data
clf.fit(X_train,Y_train)

# Extract model weights (coefficients of each feature)
w = clf.coef_

# Count and print the number of non-zero coefficients
# (number of features that the model kept)
print(np.count_nonzero(w))
```

```python
import numpy as np

clf= LogisticRegression(C=0.01, penalty='l1', solver='liblinear')
clf.fit(X_train,Y_train)

w = clf.coef_

print(np.count_nonzero(w))
```
