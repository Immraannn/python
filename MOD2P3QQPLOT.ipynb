```python
#Q-Q plot
import numpy as np 
import pylab 
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
# N(0,1)
std_normal = np.random.normal(loc = 0, scale = 1, size=1000)# loc=mean,scale=std-dev

# 0 to 100th percentiles of std-normal
for i in range(0,101):
    print(i, np.percentile(std_normal,i))


```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# N(0,1)
std_normal = np.random.normal(loc=0, scale=1, size=1000)

# Plot using seaborn
#sns.kdeplot(std_normal, fill=True)
sns.histplot(std_normal, kde=True,stat="density",  bins=130)#stat="density" changes the y-axis to show probability density, NOT counts.
                                                             #bins for number of bar

plt.title("Standard Normal Distribution (N(0,1))")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# generate 100 sanples from N(20,5)
measurements = np.random.normal(loc = 0, scale = 1, size=100) 
plt.xlim(-1,1)# Set x-axis limits from -1 to +1 (only show central region of plot)
stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```

```python
# Generate 100 random samples from a Normal distribution N(20, 5^2)
# loc = mean = 20
# scale = standard deviation = 5
# size = number of samples = 100
measurements = np.random.normal(loc=20, scale=5, size=100)

# Create a Q-Q plot to check if data follows a normal distribution
# dist = "norm" means compare data with theoretical normal distribution
stats.probplot(measurements, dist="norm", plot=pylab)

# Show the plot
pylab.show()
```
