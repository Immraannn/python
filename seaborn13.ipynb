```python
import seaborn  as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=sns.load_dataset('tips')
data.head()
```

```python
sns.relplot(x='total_bill',y='tip',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',data=data,kind='line')

```

```python
#hue PARAMETER
sns.relplot(x='total_bill',y='tip',hue='smoker',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',sizes=(115,200),data=data)
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=data)
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
plt.style.use('fivethirtyeight')
sns.catplot(x='day',y='tip',kind='strip',data=tips)
```

```python
sns.catplot(x='day',y='tip',kind='strip',jitter=1.2,data=tips)
```

```python
sns.catplot(x='day',y='tip',kind='strip',jitter=0,data=tips)
```

```python
sns.catplot(x='day',y='tip',kind='swarm',data=tips)
```

```python
sns.catplot(x='day',y='tip',kind='swarm',hue='sex',data=tips)
```

```python
sns.swarmplot(x='day',y='tip',hue='sex',data=tips)
```

```python
#BOXPLOT
#UNIVARIATE
sns.boxplot(tips['tip'])
```

```python
sns.boxplot(tips['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='box',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',kind='box',data=tips,hue='sex')
```

```python
#violin plot
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=tips,hue='sex')
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=tips,hue='sex',split=True)
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',data=tips)#dash in output show mean confidence interval
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',data=tips,hue='sex')
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',data=tips,hue='sex',estimator=np.median)#default is mean
```

```python
sns.catplot(x='sex',kind='count',data=tips)
```

```python
sns.catplot(x='sex',kind='count',data=tips,hue='smoker')
```
