```python
import pandas as pd
india_weather=pd.DataFrame({'city':['mumbai','delhi','bangalore'],
                            'temperature':[32,45,30],
                            'humidity':[80,60,78]})
india_weather
```

```python

us_weather=pd.DataFrame({'city':['newyork','chicago','orlando'],
                          'temperature':[21,14,35],
                          'humidity':[68,65,75]})
us_weather
```

```python
#concate two dataframe
df=pd.concat([india_weather,us_weather])
df
```

```python
df=pd.concat([india_weather,us_weather],ignore_index=True)
df
```

```python
df=pd.concat([india_weather,us_weather],axis=1)
df
```

```python
#MERGE DATAFRAME
```

```python
temp_df=pd.DataFrame({'city':['newyork','chicago','orlando','texas'],
                          'temperature':[21,14,35,42]})
temp_df
```

```python
humid_df=pd.DataFrame({'city':['newyork','chicago','orlando'],
                          'humidity':[68,65,75]})
humid_df
```

```python
df=pd.merge(temp_df,humid_df,on='city')
df
```

```python
#OUTER JOIN
df=pd.merge(temp_df,humid_df,on='city',how='outer')
df
```
