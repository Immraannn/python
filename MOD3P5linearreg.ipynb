```python
from sklearn.datasets import load_diabetes
df = load_diabetes()

print(df.feature_names)# prints the column (feature) names
print(df.data[:5])# prints the first 5 rows of input features (X)
print(df.target[:5])# prints the first 5 target values (y)
print(df.data.shape)
```

```python
import pandas as pd

# df1: Create a new pandas DataFrame using the raw feature data from the sklearn dataset
df1 = pd.DataFrame(df.data)   # df.data â†’ only the numerical feature values (no column names by default)

# head(): Shows the first 5 rows of the DataFrame
print(df1.head())
#df.head()      # shows first 5 rows
#df.head(10)    # shows first 10 rows
```

```python
df1 = pd.DataFrame(df.data, columns=df.feature_names)
print(df1.head())
```

```python
# Add the target (house prices) as a new column named "PRICE" in df1 DataFrame
df1['PRICE'] = df.target   # df.target contains the output values (y)

# Separate input features (X) and target output (Y)
print(df1.head(1))
# X will contain all columns except PRICE (i.e., model inputs)
X = df1.drop('PRICE', axis = 1)   # axis=1 means drop a column
print(X.head(1))
# Y will contain only the PRICE column (i.e., model output / target)
Y = df1['PRICE']
print(Y.head(3))
```

```python
from sklearn.model_selection import train_test_split

# Split the dataset
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size = 0.33, random_state = 5
)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression   # import Linear Regression model
import matplotlib.pyplot as plt                    # import library for plotting

lm = LinearRegression()     # create a Linear Regression model object

lm.fit(X_train, Y_train)    # train (fit) the model using training data
                           # X_train â†’ input features
                           # Y_train â†’ actual prices

Y_pred = lm.predict(X_test) # predict the prices for test data (X_test)
                           # model outputs predicted values â†’ Y_pred
import numpy as np
# Plot regression line (perfect prediction line y = x)

# np.linspace(start, end, number_of_points)
# This creates 100 evenly spaced points between the minimum & maximum actual prices (Y_test)
# Basically, it generates points for a straight reference line (y = x)
line = np.linspace(min(Y_test), max(Y_test), 100)


# This draws a 45-degree perfect regression line (y = x)
# If the model was perfect, all predicted points would lie exactly on this line
plt.plot(line, line)  # regression reference line

# Plot actual vs predicted prices
plt.scatter(Y_test, Y_pred)   # scatter plot between actual prices and predicted prices
plt.xlabel("Prices: $Y_i$")   # x-axis label (actual prices)
plt.ylabel("Predicted prices: $\hat{Y}_i$")   # y-axis label (predicted prices)
plt.title("Prices vs Predicted prices: $Y_i$ vs $\hat{Y}_i$") # graph title

plt.show()   # display the graph
```

```python
# Calculate prediction error (difference between actual & predicted values)
delta_y = Y_test - Y_pred  

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Set a simple plot style
sns.set_style("whitegrid")

# Plot the KDE (smooth distribution) of the errors
sns.kdeplot(delta_y, bw_adjust=0.5)   # bw_adjust controls smoothness of curve

plt.xlabel("Error (Y_test âˆ’ Y_pred)")  # x-axis label
plt.title("Distribution of Prediction Errors")  # plot title
plt.show()  # display plot
```

```python
sns.set_style('whitegrid')
sns.kdeplot(np.array(Y_pred), bw=0.5)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# âœ… Calculate predictions (already done earlier)
# Y_pred = lm.predict(X_test)

# âœ… Calculate residuals (errors)
residuals = Y_test - Y_pred

# âœ… Evaluation Metrics
mse = mean_squared_error(Y_test, Y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(Y_test, Y_pred)

print("Model Performance Metrics:")
print(f"RÂ² Score: {r2:.4f}")     # closer to 1 = better
print(f"MSE: {mse:.4f}")         # mean squared error
print(f"RMSE: {rmse:.4f}")       # root mean squared error

# âœ… Residual Plot (Actual vs Error)
plt.figure(figsize=(6,4))
plt.scatter(Y_pred, residuals)
plt.axhline(0)  # horizontal reference line at zero
plt.xlabel("Predicted Values")
plt.ylabel("Residuals (Actual - Predicted)")
plt.title("Residual Plot")
plt.show()

# âœ… KDE curve of residuals (error distribution)
sns.set_style("whitegrid")
sns.kdeplot(residuals, bw_adjust=0.5)
plt.xlabel("Residual Error")
plt.title("Distribution of Residuals")
plt.show()
```

```python
from sklearn.metrics import r2_score

# Predict on train & test
lm= LinearRegression()
lm.fit(X_train, Y_train)
y_pred_train = lm.predict(X_train)
y_pred_test = lm.predict(X_test)

# Print RÂ² scores
print("Train R2 Score:", r2_score(Y_train, y_pred_train))
print("Test R2 Score :", r2_score(Y_test, y_pred_test))
```
