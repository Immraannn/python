```python
# Import required libraries
!pip install mlxtend
from sklearn import neighbors                    # For KNN classifier
from mlxtend.plotting import plot_decision_regions  # For plotting decision boundaries
import matplotlib.pyplot as plt                  # For visualization

# Function definition: performs KNN & plots decision boundary
def knn_comparision(data, k):
    
    # Extract feature columns x1 and x2 from DataFrame as NumPy array
    # X becomes the input features used for KNN
    X = data[['x1','x2']].values

    # Extract target output column 'y', convert to int, and convert to NumPy array
    # y stores the class labels
    y = data['y'].astype(int).values

    # Initialize the KNN classifier from sklearn
    # n_neighbors = k means model will look at k nearest neighbors
    clf = neighbors.KNeighborsClassifier(n_neighbors=k)

    # Train the model on given features X and labels y
    clf.fit(X, y)

    # Plot decision boundary using the trained model
    # X -> feature data, y -> actual class labels, clf -> trained model
    # legend=2 adds legend for classes
    plot_decision_regions(X, y, clf=clf, legend=2)

    # Label X-axis as X1
    plt.xlabel('X1')

    # Label Y-axis as X2
    plt.ylabel('X2')

    # Set title of graph, including value of k used in KNN
    plt.title('Knn with K='+ str(k))

    # Display the plot
    plt.show()
```

```python
df=pd.read_csv('6.overlap.csv')
print(df.head(5))

data = pd.read_csv('6.overlap.csv', names=['x1', 'x2', 'y'])
print(data.head(5))
for i in [1, 5, 15, 30, 45]:
    knn_comparision(data, i)
```
